
# Scenario & Sensitivity Analysis – Business Case

## Objective
Analyze how **price, cost, and demand** impact revenue and profit under different scenarios.

## Dataset
Synthetic dataset representing a product business:
- Price per unit
- Cost per unit
- Demand (units sold)

We simulate:
- Base Case
- Best Case
- Worst Case


```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```

# Base dataset
data = pd.DataFrame({
    'Scenario': ['Base'],
    'Price': [100],
    'Cost': [60],
    'Demand': [1000]
})

# Function to calculate metrics
def calculate_metrics(df):
    df['Revenue'] = df['Price'] * df['Demand']
    df['Total Cost'] = df['Cost'] * df['Demand']
    df['Profit'] = df['Revenue'] - df['Total Cost']
    df['Profit Margin'] = df['Profit'] / df['Revenue']
    return df

data = calculate_metrics(data)
data
```

```

# Create scenarios
scenarios = pd.DataFrame({
    'Scenario': ['Base', 'Best Case', 'Worst Case'],
    'Price': [100, 120, 90],
    'Cost': [60, 55, 70],
    'Demand': [1000, 1200, 800]
})

scenarios = calculate_metrics(scenarios)
scenarios
```

```

# Profit comparison
plt.figure()
plt.bar(scenarios['Scenario'], scenarios['Profit'])
plt.title('Profit Comparison Across Scenarios')
plt.xlabel('Scenario')
plt.ylabel('Profit')
plt.show()
```

```

# Sensitivity analysis on price
price_range = np.arange(80, 140, 5)

sensitivity = pd.DataFrame({
    'Price': price_range
})

sensitivity['Demand'] = 1000
sensitivity['Cost'] = 60

sensitivity = calculate_metrics(sensitivity)

plt.figure()
plt.plot(sensitivity['Price'], sensitivity['Profit'])
plt.title('Sensitivity Analysis: Price vs Profit')
plt.xlabel('Price')
plt.ylabel('Profit')
plt.show()
```


## Risk vs Reward Analysis

- **Best Case**: High reward, lower risk due to cost optimization and demand increase
- **Worst Case**: High risk due to rising costs and falling demand
- **Base Case**: Stable but moderate returns

## Key Insight
Profit is highly sensitive to price changes, followed by demand, then cost.
